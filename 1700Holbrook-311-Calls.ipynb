# 311 Customer Service Requests at 1700 Holbrook

```python
import json
import pandas as pd

with open('311_Customer_Service_Requests_2026.geojson') as f:
    data = json.load(f)

geojson = pd.json_normalize(data["features"])

def write_to_csv(filename, df):
    file = f'{filename}.csv'
    return df.to_csv(file, index=False)

geojson = geojson.filter(
    items=['properties.MethodReceived','properties.CreatedDate','properties.SRStatus','properties.StatusDate','properties.DueDate','properties.CloseDate','properties.Agency','properties.LastActivity','properties.LastActivityDate','properties.Outcome','properties.Address','properties.ZipCode','properties.Neighborhood'])


write_to_csv("311_Customer_Service_Requests_2026", geojson)
```
